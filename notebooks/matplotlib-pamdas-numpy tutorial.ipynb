# Intro to numpy

```python
# Creating 1D arrays
arr1d = np.array([1,2,4,5])
print(arr1d)
```

```python
# Creating 2D arrays
arr2d = np.array([[1,2,4,5],[10,39,57,3]])
print(arr2d)
```

```python
# Creating 3D arrays

arr3d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 67, 3], [4, 10, 12]]])

print(arr3d)
```

```python
print(arr3d[0][1][2]) # get 6
```

```python
# Check dims
print(arr1d.ndim)
print(arr2d.ndim)
print(arr3d.ndim)
```

```python
# Indexing into 1D arrays. Get 4 from arr1d
print(arr1d[2])
```

```python
# Indexing into 2D arrays. Get 57 from arr2d
print(arr2d[1][2])
```

```python
# Indexing into 3D arrays. Get 67 from arr3d
print(arr3d[1][0][1])
```

```python
# Slicing 1D arrays
arr = np.array([1, 2, 3, 4, 5, 6, 7])
print(arr)
```

```python
print(arr[4:]) # get elements from 4th index to the end of the array
```

```python
print(arr[:4]) # get elements from the beginning of the array to the 4th index
```

```python
#arr = np.array([1, 2, 3, 4, 5, 6, 7])
#idx = np.array([-6, -5, -4, -3, -2, -1, 0])

print(arr[-3:])
```

```python
print(arr[-3:-1])
```
